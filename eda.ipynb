```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20,10)
```

```python
LABEL_F = '<YOUR FILE>'
```

```python
labels = open(LABEL_F).readlines()
labels = [float(labels[i].split()[0]) for i in range(len(labels))]
labels = [round(label) for label in labels]
labels = np.array(labels)
```

```python
min(labels)
```

```python
max(labels)
```

```python
tick = np.arange(0, len(labels), 1000)
plt.xticks(tick)
plt.plot(labels);
```

```python
tick = np.arange(0, 29)
plt.xticks(tick)
plt.hist(labels, 29);
```
